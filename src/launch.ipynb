```python
import subprocess 
import os 
import sys

currentDir = os.getcwd() # get pull path of current location
project_root = os.path.join(currentDir, "..") # go one folder back in the path  
if project_root not in sys.path: # add this path to folder that ipynb will search for moduls
    sys.path.append(project_root)

from tools.checker import checker
from tools.warehouse_loader import WarehouseLoader
from tools.warehouse_loader import WarehouseInstance

# num -> instance
# 0 : toy
# 1 : big_category
# 2 : big_family
# 3 : big_market

def execCode(instance): 
    """ Execute C++ code on wanted instance number """
    cmd = ["./prog", str(instance)] # command to launch
    subprocess.run(cmd, text=True, check=True)

def execAndCheck(instance): 
    """ Execute instance number & check validity of solution obtained """

    if instance !=0 and instance != 1 and instance != 2 and instance != 3: 
        raise Exception("nom instance incorect")
    
    if instance == 0: 
        print("Launching toy")
        warehouseDir = "../data/warehouse_toy"
    if instance == 1: 
        print("Launching big cat")
        warehouseDir = "../data/warehouse_big_/warehouse_big_category"
    if instance == 2: 
        print("Launching big fam")
        warehouseDir = "../data/warehouse_big_/warehouse_big_family"
    if instance == 3: 
        print("Launching big mar")
        warehouseDir = "../data/warehouse_big_/warehouse_big_market"

    execCode(instance)

    # checking solution writen
    loader = WarehouseLoader(warehouseDir) 
    inst = loader.load_all()

    print("=== INSTANCE ===") # print inst metadata 
    m = inst.metadata
    print(f"Racks: {m['num_racks']} | Produits: {m['num_products']} | "
          f"Circuits: {m['num_circuits']} | Commandes: {m['num_orders']}")
    
    print("\n=== VALIDATION ===") # print validation status 
    solution_dir = os.path.join(warehouseDir, "solutions") # get path to wanted solution
    checker(inst, solution_dir) # call checker from tools 

    
execAndCheck(3)

```
